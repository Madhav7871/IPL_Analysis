### Variables Description

AVG: Batting Average (Total Runs divided by the number of times they were dismissed).

SR: Strike Rate (Average number of runs scored per 100 balls faced).

50s: Total number of half-centuries (scores between 50 and 99) hit.

6s: Total number of sixes hit by the player.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
dataset['HS'] = dataset['HS'].astype(str).str.replace('*', '')
dataset['HS'] = pd.to_numeric(dataset['HS'])

dataset['AVG'] = dataset['AVG'].astype(str).str.replace('-', '0')
dataset['AVG'] = pd.to_numeric(dataset['AVG'])

dataset.drop_duplicates(inplace=True)
````

### What all manipulations have you done and insights you found?

Highest Score (HS) Standardization: The raw HS column contained an asterisk * symbol to indicate that a batter remained not out during their highest-scoring innings. This forced the column to be treated as text (object). I stripped the * characters and converted the entire column into an integer numerical type so it can be used for mathematical operations and plotting.

Batting Average (AVG) Correction: Some batsmen did not get dismissed or did not bat enough to have a calculated average, resulting in a hyphen (-) in the raw data. I replaced these hyphens with 0 and converted the column to a float numerical type to ensure smooth computations and prevent runtime errors during plotting.

Deduplication: A duplicate check was executed across the entire dataset, and drop_duplicates() was run to guarantee that every row uniquely represents a single player with no redundant entries.

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1 visualization code
import matplotlib.pyplot as plt
import seaborn as sns

# Setting a global dark background style
plt.style.use('dark_background')

# Initialize figure and set custom dark facecolors for a modern UI look
fig, ax = plt.subplots(figsize=(14, 7))
fig.patch.set_facecolor('#121212')
ax.set_facecolor('#121212')

# Creating a histogram with KDE using high-energy neon colors
sns.histplot(dataset['Runs'], bins=25, kde=True,
             color='#00E5FF', edgecolor='#121212', alpha=0.7,
             line_kws={'linewidth': 3, 'color': '#FF007F'}, ax=ax)

# Enhancing aesthetics and labels with distinct white/grey tones
plt.title('Univariate Analysis: Distribution of Total Runs Scored (IPL 2025)', fontsize=18, fontweight='bold', color='white', pad=15)
plt.xlabel('Total Runs Scored', fontsize=14, fontweight='bold', color='#E0E0E0', labelpad=10)
plt.ylabel('Frequency (Number of Players)', fontsize=14, fontweight='bold', color='#E0E0E0', labelpad=10)

# Adding subtle gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.2, color='white')

# Refining the axes borders
sns.despine(ax=ax, top=True, right=True)
ax.spines['left'].set_color('#333333')
ax.spines['bottom'].set_color('#333333')
ax.tick_params(colors='#E0E0E0')

plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A Histogram paired with a Kernel Density Estimate (KDE) curve is the absolute best chart for univariate continuous data. It allows us to immediately see the shape, spread, and skewness of the runs scored across the entire dataset of 156 players, showing exactly where the majority of players fall.

##### 2. What is/are the insight(s) found from the chart?

The distribution is heavily right-skewed (a long tail towards the right). The massive peak on the left indicates that the vast majority of players scored between 0 and 100 runs throughout the tournament (these are primarily specialist bowlers, lower-order batters, or bench players who played very few matches). Conversely, the long tail shows that players scoring 400+ runs are exceedingly rare, highlighting the scarcity of elite top-order run accumulators.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Positive Impact: Yes. Franchise owners and talent scouts can use this to understand market scarcity. Since 400+ run scorers are statistically rare, franchises can justify allocating premium auction budgets to retain or buy these consistent top-order players, ensuring steady batting performances.

Negative Growth/Risk: A key negative insight here is squad fragility. If a team's total run composition is dominated by just 1 or 2 players sitting in the far right tail, an injury to them could collapse the entire team's performance, as the vast majority of the squad statistically sits in the low-scoring bracket. Teams must ensure they have a balanced middle order rather than just relying on outliers.

#### Chart - 6

````python
# Chart - 6 visualization code
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(14, 8))
fig.patch.set_facecolor('#0a0a0a')
ax.set_facecolor('#0a0a0a')

# Grouping boundaries by team
team_boundaries = dataset.groupby('Team')[['4s', '6s']].sum().sort_values(by='6s', ascending=False)

# Stacked Bar Chart for Boundaries
team_boundaries.plot(kind='bar', stacked=True, ax=ax,
                     color=['#00e676', '#d500f9'], edgecolor='#0a0a0a', linewidth=1.5)

plt.title('Bivariate Analysis: The Boundary Breakdown (4s vs 6s by Team)', fontsize=18, fontweight='bold', color='white', pad=15)
plt.xlabel('Franchise', fontsize=14, fontweight='bold', color='#E0E0E0', labelpad=10)
plt.ylabel('Total Count of Boundaries', fontsize=14, fontweight='bold', color='#E0E0E0', labelpad=10)

plt.legend(['Fours (4s)', 'Sixes (6s)'], frameon=False, labelcolor='white', fontsize=12)
plt.grid(axis='y', linestyle='-.', alpha=0.2, color='#00e676')
sns.despine(top=True, right=True)
ax.tick_params(colors='#E0E0E0', rotation=0)

plt.tight_layout()
plt.show()
````

A Stacked Bar Chart is the best way to compare absolute totals (total boundaries) across categories (Teams) while simultaneously breaking down the composition of those totals (the ratio of Fours to Sixes).

##### 2. What is/are the insight(s) found from the chart?

We can instantly identify the tactical approach of different franchises. Teams with a massive purple block rely heavily on sheer power hitting and clearing the ropes (6s). Teams with a dominant green block prioritize finding gaps, placement, and running between the wickets to accumulate boundaries (4s).

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Positive Impact: This dictates home-ground curation strategy. A team heavily reliant on 6s should curate flat pitches with shorter boundaries at their home stadium to maximize their specific squad strength, directly leading to more home wins.

Negative Growth/Risk: A team with a very low six-hitting ratio will struggle massively in high-pressure run chases where the required run rate spikes. Inability to hit sixes in modern T20s is a severe tactical handicap that leads to tournament elimination.

#### Chart - 7

````python
# Chart - 7 visualization code
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(14, 8))
fig.patch.set_facecolor('#181824')
ax.set_facecolor('#181824')

# Boxplot to check batting average consistency across teams
sns.boxplot(x='Team', y='AVG', data=dataset, palette='Set3',
            boxprops=dict(alpha=0.7, edgecolor='white', linewidth=2),
            whiskerprops=dict(color='white', linewidth=1.5),
            capprops=dict(color='white', linewidth=1.5),
            medianprops=dict(color='#ff0055', linewidth=3), ax=ax)

# Adding a stripplot to show individual data points on top of the boxes
sns.stripplot(x='Team', y='AVG', data=dataset, color='#00e5ff', alpha=0.6, size=6, jitter=True, ax=ax)

plt.title('Bivariate Analysis: Batting Average Consistency by Team', fontsize=18, fontweight='bold', color='white', pad=15)
plt.xlabel('Team', fontsize=14, fontweight='bold', color='#E0E0E0', labelpad=10)
plt.ylabel('Batting Average', fontsize=14, fontweight='bold', color='#E0E0E0', labelpad=10)

plt.grid(axis='y', linestyle='--', alpha=0.2, color='white')
sns.despine(top=True, right=True)
ax.tick_params(colors='#E0E0E0')

plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A Boxplot overlaid with a Strip Plot provides a comprehensive view of categorical data distributions. The boxplot shows the median, interquartile range (IQR - the middle 50% of players), and spread, while the strip plot reveals the exact position of every single player's average, making it easy to spot outliers.

##### 2. What is/are the insight(s) found from the chart?

We can observe which teams possess a consistent batting lineup versus heavily reliant ones. A compact box indicates that most batters in that team have similar averages (consistency). A very tall box, or a box with high floating points (outliers), shows a team relying on one or two star players while the rest of the lineup fails to contribute meaningfully.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Positive Impact: Teams with high medians and tight boxes demonstrate excellent squad depth. This proves that their scouting and player development strategies are highly successful and should be replicated in future seasons.

#### Chart - 8

````python
# Chart - 8 visualization code
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(12, 9))
fig.patch.set_facecolor('#050510')
ax.set_facecolor('#050510')

# 2D Kernel Density Estimate (Contour / Topographic Plot)
# This maps the "gravity center" of player performances
sns.kdeplot(x=dataset['SR'], y=dataset['AVG'], cmap='mako',
            fill=True, thresh=0.05, levels=12, alpha=0.9, ax=ax)

# Layering a subtle scatter plot on top to see the exact player positions
sns.scatterplot(x=dataset['SR'], y=dataset['AVG'], color='#00ffcc',
                s=30, alpha=0.6, edgecolor='white', ax=ax)

plt.title('Bivariate Analysis: Performance Density Contour (AVG vs SR)', fontsize=18, fontweight='bold', color='white', pad=15)
plt.xlabel('Strike Rate (Pace of Scoring)', fontsize=14, fontweight='bold', color='#E0E0E0', labelpad=10)
plt.ylabel('Batting Average (Consistency)', fontsize=14, fontweight='bold', color='#E0E0E0', labelpad=10)

# Highlighting the core density zone
plt.axvline(x=130, color='#ff007f', linestyle='--', alpha=0.4)
plt.axvline(x=160, color='#ff007f', linestyle='--', alpha=0.4)
plt.text(135, dataset['AVG'].max() - 5, 'CORE PLAYER\nDENSITY ZONE', color='#ff007f', fontweight='bold', fontsize=12)

plt.grid(True, linestyle='-', alpha=0.05, color='white')
sns.despine(top=True, right=True)
ax.tick_params(colors='#E0E0E0')

plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A 2D Density Contour Plot (KDE) acts like a topographic map. Instead of just showing scattered dots where data overlaps and becomes unreadable, it uses glowing color gradients to show the exact "gravity center" where the highest density of players sits. Anyone mapped outside this glowing contour core—especially towards the top right—is statistically a rare outlier.

````python
# Chart - 9 visualization code
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(14, 8))
fig.patch.set_facecolor('#120a17')
ax.set_facecolor('#120a17')

# Feature Engineering: Calculating Runs from Boundaries vs Running
dataset_calc = dataset.copy()
dataset_calc['Boundary_Runs'] = (dataset_calc['4s'] * 4) + (dataset_calc['6s'] * 6)
dataset_calc['Running_Runs'] = dataset_calc['Runs'] - dataset_calc['Boundary_Runs']

# Getting top 10 highest run scorers
top_10 = dataset_calc.nlargest(10, 'Runs')

# Stacked Bar for Run Composition
ax.bar(top_10['Player Name'], top_10['Boundary_Runs'], label='Runs from Boundaries (4s & 6s)', color='#ff0055', edgecolor='white')
ax.bar(top_10['Player Name'], top_10['Running_Runs'], bottom=top_10['Boundary_Runs'], label='Runs from Running (1s, 2s, 3s)', color='#00e5ff', edgecolor='white')

plt.title('Multivariate Analysis: Run Composition of Top 10 Batters (Boundaries vs Running)', fontsize=18, fontweight='bold', color='white', pad=15)
plt.xlabel('Player Name', fontsize=14, fontweight='bold', color='#E0E0E0', labelpad=10)
plt.ylabel('Total Runs', fontsize=14, fontweight='bold', color='#E0E0E0', labelpad=10)

plt.xticks(rotation=45, ha='right', color='#E0E0E0', fontweight='bold')
plt.yticks(color='#E0E0E0')
ax.legend(frameon=False, labelcolor='white', fontsize=12, loc='upper right')

plt.grid(axis='y', linestyle=':', alpha=0.3, color='white')
sns.despine(top=True, right=True)

plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A Composition Stacked Bar Chart is perfect for taking a single total metric (Total Runs) and breaking it down into its constituent parts. By feature engineering the data to separate boundary runs from running runs, this chart uniquely visuals how a player achieves their score, not just what they scored.

##### 2. What is/are the insight(s) found from the chart?

This chart profiles a player's physical fitness and batting style. A massive pink section indicates a pure power-hitter who relies on clearing the ropes. A large cyan section indicates a highly fit, agile player who excels at gap-placement and strike rotation (running hard between the wickets).

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Positive Impact: This dictates opening partnership strategies. A team can use this to perfectly pair a pure power-hitter (huge pink bar) with a strike-rotator (huge cyan bar). This ensures the scoreboard keeps moving regardless of whether boundaries are being hit, maximizing the total team score.

Negative Growth/Risk: If a player has a huge boundary run count but almost zero running runs, it implies poor physical fitness or an inability to rotate the strike against spin bowling. On large grounds with long boundaries, this player will become a massive liability, stalling the innings and creating negative momentum.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(12, 10))
fig.patch.set_facecolor('#000000')
ax.set_facecolor('#000000')

# Selecting only numerical columns for correlation
numerical_data = dataset.select_dtypes(include=[np.number])
correlation_matrix = numerical_data.corr()

# Creating a mask to hide the upper triangle for a cleaner look
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Heatmap with an intense, dark-themed colormap
sns.heatmap(correlation_matrix, mask=mask, annot=True, fmt=".2f",
            cmap='inferno', vmin=-1, vmax=1,
            cbar_kws={"shrink": .8}, linewidths=1, linecolor='#222222', ax=ax)

plt.title('Multivariate Analysis: Correlation Matrix of Batting Metrics', fontsize=18, fontweight='bold', color='white', pad=20)

plt.xticks(rotation=45, ha='right', color='#E0E0E0', fontweight='bold')
plt.yticks(rotation=0, color='#E0E0E0', fontweight='bold')

plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A Correlation Heatmap is the absolute standard for Multivariate Analysis when attempting to find relationships across all numerical variables at once. It uses color intensity to represent the strength and direction of the linear relationships, making complex multi-variable interactions instantly readable.

##### 2. What is/are the insight(s) found from the chart?

The heatmap reveals exactly which metrics drive run-scoring. For example, we typically see a near-perfect positive correlation between Runs and Balls Faced (BF), and very strong correlations between Runs and boundaries (4s, 6s). Interestingly, we can observe whether Strike Rate (SR) has any correlation with Average (AVG), which often reveals if aggressive players tend to get out faster.
